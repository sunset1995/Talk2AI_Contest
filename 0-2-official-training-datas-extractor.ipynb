```python
import time
import re
import os

import jieba
jieba.set_dictionary('datas/dict/dict.txt.big')
jieba.load_userdict('datas/dict/edu_dict.txt')
```

```python
corpus = {}
for corpus_dir in ['datas/training_data/subtitle_with_TC', 'datas/training_data/subtitle_no_TC']:
    not_chinese_word = u'[^\u4e00-\u9fff]'  # only keep Chinese word, remove 
    not_chinese_word_rule = re.compile(not_chinese_word)
    for dir_name in os.listdir(path=corpus_dir):
        if dir_name not in corpus:
            corpus[dir_name] = []
        full_dir_name = os.path.join(corpus_dir, dir_name)
        for f_name in os.listdir(path=full_dir_name):
            full_name = os.path.join(full_dir_name, f_name)
            if not os.path.isfile(full_name):
                continue
            with open(full_name) as f:
                corpus[dir_name].append([not_chinese_word_rule.sub(' ', line).strip() for line in f])
                corpus[dir_name][-1] = [s for s in corpus[dir_name][-1] if len(s) > 0]
            print('%4d lines in %s' % (len(corpus[dir_name][-1]), full_name))
```

```python
start_time = time.time()
rule = re.compile('\s+')
for k, v in corpus.items():
    f_name = os.path.join('datas/training_data/', k) + '.txt'
    with open(f_name, 'w') as fo:
        fo.write('\n'.join(['\t'.join([rule.sub(' ', ' '.join(list(jieba.cut(s)))) for s in cp]) for cp in v]))
    print('Export: %40s / elapsed time %10.f sec.' % (f_name, time.time() - start_time))
```

```python
corpus_fnames = [
    'datas/training_data/no_TC_下課花路米.txt',
    'datas/training_data/no_TC_誰來晚餐.txt',
    'datas/training_data/no_TC_公視藝文大道.txt',
    'datas/training_data/no_TC_成語賽恩思.txt',
    'datas/training_data/no_TC_我的這一班.txt',
    'datas/training_data/no_TC_流言追追追.txt',
    'datas/training_data/no_TC_人生劇展.txt',
    'datas/training_data/no_TC_聽聽看.txt',
]
```

```python
threshold = 17
for fname in corpus_fnames:
    with open(fname, 'r') as f:
        lst = [line.strip().split('\t') for line in f]
        n_lst = []
        for episode in lst:
            if max([len(s.strip().split()) for s in episode]) > threshold:
                continue
            n_lst.append(episode)
        with open(fname, 'w') as fo:
            fo.write('\n'.join(['\t'.join([s.strip() for s in episode]) for episode in n_lst]))
```

```python
max_len_lst = []
for fname in corpus_fnames:
    with open(fname, 'r') as f:
        lst = [line.strip().split('\t') for line in f]
        for episode in lst:
            max_len_lst.append(max([len(s.strip().split()) for s in episode]))
max_len_lst.sort(reverse=True)
max_len_lst[:100]
```
